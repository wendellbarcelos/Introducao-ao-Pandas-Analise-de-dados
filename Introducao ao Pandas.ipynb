## Criando um dataframe a partir de um dicionário

```python
# Formas de criar um Dataframe seria
# dataframe = pd.dataframe()  essa é a forma de criar um dataframe porem esta vazio

vendas = {'data': ['21/02/2022', '26/02/2022'],
          'valor': [500, 300],
          'Produto': ['Feijão', 'Arroz'],
          'Qtd': [300, 200]
         }
# Temos um dicionario com os valores na qual precisamos usar o pd.dataframe para transforma-lo em um df

vendas_df = pd.DataFrame(vendas)  ## Dessa forma criamos um dataframe usando o dicionario
```

## Visualização dos Dados

## Importando arquivos e bases de dados

```python
# Vamos importar os dados de um arquivo do excel que não esta na mesma pasta do projeto
dados = pd.read_excel('./dados/Vendas.xlsx')
```

## Resumos de Visualização de Dados simples e úteis

```python
display(dados.head(10))
print(dados.shape)
display(dados.describe())
```

## Pegar 1 coluna (e os pd.Series)

```python
# Quando retornamos uma coluna do dataframe ela vem como uma serie e nao como um dataframe
display(dados['Produto'])

# Para retornar uma coluna como um dataframe precisamos colocar dentro de dois colchetes, isso vai gerar uma tabela
display(dados[['Produto']])
```

## .loc, um método muito importante

 - Pegar por linha
 - Pegar linha de acordo a alguma condição
 - Pegar linhas e colunas especificas
 - Pegar 1 valor especifico

```python
# Pegar por linha
display(dados.loc[2])  # Dessa forma ira retornar uma serie da linha com o indice 2, também pode ser usado fatiamento

# Pegar linha de acordo a alguma condição
display(dados.loc[dados['ID Loja'] == 'Norte Shopping'])  # Retornará todos os valores cujo o ID Loja seja Norte Shopping

# Pegar linhas e colunas especificas
display(dados.loc[dados['ID Loja'] == 'Norte Shopping', ['ID Loja', 'Produto', 'Quantidade']]) # Retorna as colunas Produto e Qtd baseada na condição

# Pegar 1 valor especifico
display(dados.loc[1:5, 'Produto']) # Retornará os produtos da linha 1 até a 5
```

## Adicionar 1 coluna

```python
# A partir de uma coluna que existe
dados['Comissão'] = dados['Valor Final'] * 0.05  # Dessa forma criamos uma coluna comissão baseado no valor final


# Criar uma coluna com valor padrão
dados.loc[:, 'Imposto'] = 0  # Com o loc vc vai passar por todas as linhas adicionando o 0 na coluna Imposto 
dados['Imposto'] = dados['Valor Final'] * 0.20  # Vamos substituir o valor de 0 da coluna Imposto por 20% do valor do produto
dados
```

## Adicionar 1 linha

```python
# Vamos carregar o arquivo com as vendas de dezembro
vendas_dez = pd.read_excel('./dados/Vendas - Dez.xlsx')

# Agora vamos usar esses dados e adicionar aos dados do dataframe anterior

dados_vendas = dados.append(vendas_dez)  # Criando uma nova variavel armazenamos os valores de dados adicionando os valores de vendas_dez
dados_vendas  # Retornará uma tabela com todos os valores adicionados
```

## Excluir linhas e colunas

```python
# Vimos que nas colunas Comissão e Imposto os valores da tabela vendas_dez estão nulos, temos varias formas de fazer
dados_vendas = dados_vendas.drop('Imposto', axis=1)
```

```python
# A coluna Imposto foi apagada da tabela 
dados_vendas
```

## Valores Vazios
 - Deletar linhas/colunas vazias
 - Deletar linhas que possuem valores vazios
 - Preencher valores vazios (média e ultimo valor)

```python
# Usando o dropna para remover os valores vazios da tabela
# dados_vendas = dados_vendas.dropna(how='all', axis=1)  # Dessa forma ira remover todos os valores que sejam nulos

# Deletar linhas que possuem pelo menos 1 valor vazio
# dados_vendas = dados_vendas.dropna()

# Preencher valores vazios (média e ultimo valor) - Dessa forma usaremos a media para preencher os valores nulos da coluna
dados_vendas['Comissão'] = dados_vendas['Comissão'].fillna(dados_vendas['Comissão'].mean())
dados_vendas

# Preenche os valores nulos com o ultimo valor
# dados_vendas = dados_vendas.ffill()
```

## Calcular Indicadores
 - Groupby 
 - Value Counts

```python
# Queremos saber quantas foram as transações por lojas, é simples usando o value counts
transacao_loja = dados_vendas['ID Loja'].value_counts()  # Retornará a qtd de transação por loja
display(transacao_loja)

# E como podemos obter o faturamento dos produtos?? Sim usando o groupby
faturamento_produto = dados_vendas[['Produto', 'Valor Final']].groupby('Produto').sum()  # Obtemos a soma dos produtos pelo valor
display(faturamento_produto)
```

## Mesclar 2 dataframes (Procurar informações de um dataframe emoutro)

```python
# A partir de todos os dados queremos saber quem são os gerentes de cada loja, mais nessa tabela não temos os gerente
# Iremos mesclar uma tabela de gerentes

gerente_df = pd.read_excel('./dados/Gerentes.xlsx')
gerente_df  # Todos os gerentes da empresa e suas lojas, agora vamos mesclar
```

```python
# Vamos pegar nossa tabela atual e usar a função merge
dados_vendas = dados_vendas.merge(gerente_df)  # Usando o merge ele vai mesclar baseando-se na coluna existente nas duas tabelas
dados_vendas   # Retornando a tabela principal com os gerentes em suas lojas
```

```python
dados_vendas = dados_vendas.drop(['Comissão', 'Gerente'], axis=1)
```

# 2- Pegar um panorama geral sobre a sua base de dados

```python
# Olhando nossos dados percebemos que precisamos pegar o faturamento para ter uma análise melhor sobre eles

display(tabela['Valor Final'].sum())

# Olhando o faturamente temos o valor de R$ 55.895,00, mais precisamos entender como chegou a esse valor
# Vamos pegar a quantidade de lojas que empresa possui e os produtos

# Vamos pegar a quantidade de lojas da empresa
display(len(tabela["ID Loja"].value_counts())) # A empresa tem 5 lojas

# Vamos verificar a quantidade de produtos
display(len(tabela["Produto"].value_counts()))  # Total de 6 produtos
```

# 3- Começar uma análise Top/Down

```python
# Vamos pegar o faturamento por loja 
tabela[['ID Loja', 'Valor Final']].groupby("ID Loja").sum()


# Agora podemos entender melhor, vemos que o faturamento na loja de Iguatemi Campinas está muito acima dos outros valores
# precisamos entender o que esta acontecendo para o faturamente está tão alto e levar para as outras lojas.
```

# 4- Entrar no detalhe para entender

```python
# Agora vamos trazer os produtos
tabela[['ID Loja', 'Produto', 'Valor Final']].groupby(['ID Loja', 'Produto']).sum()


# Agrupando as lojas e os produtos vamos somar a coluna valor final e temos o resultado de faturamento por produto de cada loja
# E podemos perceber que em Iguatemi Campinas tem um produto que não temos nas outras lojas na qual está gerando um otimo faturamento
```

Podemos ver que na loja de Iguatemi Campinas tem um produto extra que esta gerando um faturamento alto.
